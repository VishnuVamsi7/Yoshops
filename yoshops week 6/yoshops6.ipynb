```python
dataframe1['Chennai'].sum()
dataframe1['Aurangabad'].sum()
dataframe2['Chennai'].sum()
dataframe2['Aurangabad'].sum()
import matplotlib.pyplot as plt

# define the values for each variable
variable1= dataframe1['Chennai'].sum()
variable2= dataframe1['Aurangabad'].sum()
variable3 = dataframe2['Chennai'].sum()
variable4= dataframe2['Aurangabad'].sum()

# define the labels for each variable
labels = ['chennai(NOV)', 'Aurangabad(NOV)', 'chennai(DEC)', 'Aurangabad(DEC)']

# define the colors for each variable
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# create the outer pie chart
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie([100], colors=['white'], radius=1)

# create the inner pie charts
ax.pie([variable1, variable2], colors=[colors[0], colors[1]], radius=0.7)
ax.pie([variable3, variable4], colors=[colors[2], colors[3]], radius=0.4)

# add the labels for each variable
ax.text(0, 0, labels[0], ha='center', va='center', fontsize=14)
ax.text(0, 0.7, labels[1], ha='center', va='center', fontsize=14)
ax.text(0, 0.4, labels[2], ha='center', va='center', fontsize=14)
ax.text(0, -0.7, labels[3], ha='center', va='center', fontsize=14)

# add a title
ax.set_title('For the year 2022')

# display the chart
plt.show()
```
