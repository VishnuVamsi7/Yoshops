```python
import requests
from bs4 import BeautifulSoup
import re

def extract_contact_info(url):
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    td_tags = soup.find_all('td', class_='college_list')
    contact_info = {}

    for td_tag in td_tags:
        data = td_tag.find_next_sibling('td').get_text(strip=True)
        if ':' in data:
            key, value = data.split(':', 1)
            contact_info[key.strip()] = value.strip()
        else:
            key = td_tag.get_text(strip=True)
            contact_info[key.strip()] = data.strip()

    return contact_info

def scrape_universities(html_content, class_names):
    soup = BeautifulSoup(html_content, 'html.parser')
    universities = soup.find_all('tr', class_=class_names)

    for university in universities:
        university_link = university.find('a')
        if university_link:
            url = university_link['href']
            title = university_link['title']
            university_name = university_link.get_text()
            university = re.sub(r"\s+", "-", university_name.lstrip().lower().replace(".", " ").replace(",", ""))
            colleges = ['Arts-Science','Medical','Engineering','Law']
            for i in colleges:    
                url = f"http://www.studyguideindia.com/Colleges/{i}/{university}.html"
                contact_info = extract_contact_info(url)
                if requests.head(url).status_code == requests.codes.ok:
                    print('University Name:', university_name)
                    print('Email:', contact_info.get('E-Mail'))
                    print('Phone Number:', contact_info.get('Phone'))
                    print('Address:', contact_info.get('Address'))
                    print('URL:', url)
                    print('-' * 50)
                else:
                    pass

print('''Andhra Pradesh - AP Arunachal Pradesh - AR Assam - AS Bihar - BR Chhattisgarh - CG Goa - GA Gujarat - GJ Haryana - HR Himachal Pradesh - HP
Jharkhand - JH
Karnataka - KA
Kerala - KL
Madhya Pradesh - MP
Maharashtra - MH
Manipur - MN
Meghalaya - ML
Mizoram - MZ
Nagaland - NL
Odisha - OD
Punjab - PB
Rajasthan - RJ
Sikkim - SK
Tamil Nadu - TN
Telangana - TG
Tripura - TR
Uttar Pradesh - UP
Uttarakhand - UK
West Bengal - WB
Andaman and Nicobar Islands - AN
Chandigarh - CH
Dadra and Nagar Haveli and Daman and Diu - DN
Lakshadweep - LD
Delhi - DL
Puducherry - PY
Jammu and Kashmir - JK
Ladakh - LA
Lakshadweep - LD''')

# Fetch HTML content from the first URL
state = input("Please enter the short form of the state: ")
url = f"http://www.studyguideindia.com/Colleges/default.asp?State={state}"
response = requests.get(url)
html_content1 = response.text

# Specify the class names for universities in the first URL
class_names1 = ['alternate', 'alternate1']

# Scrape and print the universities from the first URL
scrape_universities(html_content1, class_names1)
```
